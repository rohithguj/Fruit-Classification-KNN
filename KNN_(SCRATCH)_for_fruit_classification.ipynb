```python
distances = []
test_df= stest
train_df= sef

def minkowski_distance(x1, x2, p=2):
    return np.power(np.sum(np.power(np.abs(x1 - x2), p))**(1/p), p)

df_dist = pd.DataFrame()
for _, test_row in test_df.iterrows():
    distances = []
    for _, train_row in train_df.iterrows():
        distance = minkowski_distance(test_row, train_row)
        distances.append(distance)
    dist_row = pd.Series(distances)
    df_dist = pd.concat([df_dist, dist_row], axis=1)

# Transpose the DataFrame for a more intuitive representation
df_dist = df_dist.transpose()

print("df_dist DataFrame:")
print(df_dist)
```

```python
p = 2  # Euclidean distance
k = 5  # Number of nearest neighbors to select

df_dist = pd.DataFrame()
for _, test_row in test_df.iterrows():
    distances = []
    indices = []
    for _, train_row in train_df.iterrows():
        distance = minkowski_distance(test_row, train_row, p)
        distances.append(distance)
        indices.append(train_row.name)  # Store the index of the training row
    dist_row = pd.Series(distances, index=indices)
    df_dist = pd.concat([df_dist, dist_row], axis=1)

# Transpose the DataFrame for a more intuitive representation
df_dist = df_dist.transpose()
df_nearest = df_dist.apply(lambda x: pd.Series(x.nsmallest(k).index), axis=1)

print("df_nearest DataFrame:")
print(df_nearest)
```
